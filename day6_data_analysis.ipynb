# Label Encoding

```python
# encoding = {"S": 0,
#             "C": 1,
#             "Q": 2}


def encode(row):
  if row == "S":
    return 0
  elif row == "C":
    return 1
  elif row == "Q":
    return 2
  else:
    return 3



data["Emabarked_encoded_fun"] = data["Embarked"].apply(encode)


# Ass-2
# Use Lambda function to achieve the Lable encoding
```

```python
encoding = {"S": 0,
            "C": 1,
            "Q": 2}

data["Embarked_encoded_dict"] = data["Embarked"].map(encoding)
data
```
