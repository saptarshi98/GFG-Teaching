```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.random.randint(1 , 100, size = 1000)
y = np.random.randint(1 , 100, size = 50)

print(x)
```

```python
plt.scatter(x, y)
plt.title("Original Data")
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

```python
x_normalised = (x - np.min(x))/(np.max(x) - np.min(x))
y_normalised = (y - np.min(y))/(np.max(y) - np.min(y))

x_standarised = (x - np.mean(x))/np.std(x)
y_standarised = (y - np.mean(y))/np.std(y)
```

```python
x_standarised
```

```python
plt.figure(figsize = (12, 8))
plt.subplot(1, 3, 1)
plt.scatter(x_normalised, y_normalised)
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.title("Normalised Data")

plt.subplot(1, 3, 2)
plt.scatter(x, y)
plt.title("original Data")


plt.subplot(1, 3, 3)
plt.scatter(x_standarised, y_standarised)
plt.title("Standarised Data")
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler, LabelEncoder, OneHotEncoder
import pandas as pd

d = pd.read_csv('titanic (1).csv')
d.head()
```

```python
min_max = MinMaxScaler()

std = StandardScaler()
```

```python
d = d.dropna()
```

```python
d["Age_Norm"] = min_max.fit_transform(d[["Age"]])
d["Age_std"] = std.fit_transform(d[["Age"]])
d.head()
```

```python
d.isna().sum()
```

```python
label_enc = LabelEncoder()
ohe = OneHotEncoder()


d["Embarked_LE"] = label_enc.fit_transform(d["Embarked"])
d
```

```python
ohe_df = pd.DataFrame(ohe.fit_transform(d[['Sex']]).toarray(), columns = ['Male', "Female"])
```

```python
ohe_df
```

```python
d_joined = d.join(ohe_df)
d_joined
```
